```python
# 06:48:18
def conv(s):
    timeobj = datetime.strptime(s, '%H:%M:%S').time()
    return timeobj

def time_diff(t1, t2):
    diff = (datetime.combine(date.today(), t2) - datetime.combine(date.today(), t1)).total_seconds()
    return diff if diff >= 0 else -diff
            
```

```python
def check_corr(route):
    for i in coridoors.index:
        name = coridoors['Stretch'][i]
        temp = coridoors['Effected Route'][i]
        temp = temp.replace(" ", "")
        route_name_list = temp.split(',')
        for r in route_name_list:
            if r == route:
                return True
    return False
```
