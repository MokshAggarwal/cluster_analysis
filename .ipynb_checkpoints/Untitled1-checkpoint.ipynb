```python
def check_corr(route):
    for i in coridoors.index:
        name = coridoors['Stretch'][i]
        temp = coridoors['Effected Route'][i]
        temp = temp.replace(" ", "")
        route_name_list = temp.split(',')
        for r in route_name_list:
            if r == route:
                return True
    return False


def helper(file_str):
    bus_last_trip = {}
    info = {}
    result = []

    con = sqlite3.connect(file_str)
    cur = con.cursor()

    for row in tqdm(cur.execute('SELECT * FROM vehicle_feed;')):
        bus = row[0]
        if bus not in info.keys():
            info[bus] = {}
            bus_last_trip[bus] = 0
    
    con.close()
    
    con = sqlite3.connect(file_str)
    cur = con.cursor()

    for row in tqdm(cur.execute('SELECT * FROM vehicle_feed;')):
        bus = row[0]
        route = str(int(row[4]))
        trip = row[5]
        time = row[8]
        if bus_last_trip[bus] == 0:
            bus_last_trip[bus] = trip
            tup = [time, 0]
            info[bus][trip] = tup
        else:
            if bus_last_trip[bus] != trip:
                last = bus_last_trip[bus]
                info[bus][last][1] = time
                result.append((route, last, info[bus][last][0], info[bus][last][1]))
                bus_last_trip[bus] = trip
                tup = [time, 0]
                info[bus][trip] = tup
            else:
                continue

    con.close()
    
    return result


def helper2(route_times):
    results_ = {}
    start_time = {}
    
    count = 0
    
    for tup in tqdm(route_times):
        
        if count == 0:
            print(tup)
            count += 1
        
        route = tup[0]
        
        if not check_corr(route):
            continue
        
        trip = tup[1]
        t1 = conv(tup[2])
        t2 = conv(tup[3])
        diff = time_diff(t1, t2)
        
        if(diff < 300):
            continue
            
        if trip not in results_.keys():
            results_[trip] = []
            start_time[trip] = []
            results_[trip].append(diff)
            start_time[trip].append(t1)
            
        else:
            results_[trip].append(diff)
            start_time[trip].append(t1)
            
    return results_, start_time


def helper3(res, sta):
    duration = {}
    flag = 0
    st = 6
    en = 7

    while(flag != 1):
        duration[f'{st}-{en}'] = 0
        st += 1
        en += 1
        st %= 24
        en %= 24
        if en == 7:
            flag = 1
            
    for trip in res.keys():
        trip_l = res[trip]
        start_l = sta[trip]
        for ind, dur in enumerate(trip_l):
            start = start_l[ind]
            hour = int(start.hour)
            key = f'{hour}-{hour + 1}' if hour < 23 else f'{hour}-0'
            duration[key] += dur
    
    return duration


def conv(s):
    timeobj = datetime.strptime(s, '%H:%M:%S').time()
    return timeobj


def time_diff(t1, t2):
    diff = (datetime.combine(date.today(), t2) - datetime.combine(date.today(), t1)).total_seconds()
    return diff if diff >= 0 else -diff
```
